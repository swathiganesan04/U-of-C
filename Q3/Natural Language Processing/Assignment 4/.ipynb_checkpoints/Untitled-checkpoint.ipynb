```python
import requests
import pandas as pd
import numpy as np
import regex as re
import matplotlib.pyplot as plt
import seaborn as sns

import nltk
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist


import sklearn
import pandas as pd

import wordcloud
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, HashingVectorizer, TfidfTransformer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn import metrics
```

```python
# Reading CSV file as a DataFrame
violations = pd.read_csv('../../Data/Food_Inspections.csv')
```

```python
violations.info()
```

```python
violations.sample(3)
```

```python
violations.Results.value_counts()
```

```python
# Selecting only records with Results = Fail
inspections = violations[(violations['Results'] == 'Fail') | (violations['Results'] == 'Pass')][['Results', 'Violations']]
inspections.shape
```

```python
# Checking number of missing Violations
inspections.Violations.isnull().sum()
```

```python
# Cleaning the data by dropping entries with missing Violations
inspections = inspections[inspections['Violations'].notna()]
inspections.Violations.isnull().sum()
```

```python
inspections.sample(5)
```

```python
# Extracting only regulation descriptions from the 'Violations' column
reg_descriptions = []
for violation in inspections['Violations']:
    # Using regular expression to find only the regulation descriptions
    reg_exp = r'(?<=(?<=\|\s|^)\d{1,2}\.\s)([A-Z\s()\-0-9\*&,;/:.\']+)(?=(?= - Comments:|\s\||$))'
    matches = re.findall(reg_exp, violation)
    reg_descriptions.append(matches)

# Adding extracted regulation descriptions as a new column in the DataFrame
inspections.loc[:,'Regulation_Desc'] = reg_descriptions
```

```python
# Extracting only comments from the 'Violations' column
comments = []
for violation in inspections['Violations']:
    # Using regular expression to find only the comments
    reg_exp = r'Comments:\s*(.*?)(?=\||$)'
    matches = re.findall(reg_exp, violation)
    comments.append(matches)

# Adding extracted comments as a new column in the DataFrame
inspections.loc[:,'Comments'] = comments
```

```python
inspections.sample(3)
```

```python
inspections['Comments_Tokens'] = inspections['Comments'].apply(lambda x: ' '.join(x))
inspections.sample(n=3, random_state=1)
```

```python
# Define a function to clean the text data
def clean_text(text):
    
    # Convert to lowercase
    text = text.lower()
    
    tokens = nltk.word_tokenize(text)
    
    # Remove single-character tokens (mostly punctuation)
    tokens = [token for token in tokens if len(token) > 1]

    # Remove numbers
    tokens = [token for token in tokens if not token.isnumeric()]

    # Remove punctuation
    tokens = [token for token in tokens if token.isalpha()]

    # Remove stop words and lemmatize the words
    #stop_words = set(nltk.corpus.stopwords.words('english'))   
    
    wnl = nltk.WordNetLemmatizer()
    #tokens=[wnl.lemmatize(token) for token in tokens if not token in stop_words]
    tokens=[wnl.lemmatize(token) for token in tokens]
    
    #ps = PorterStemmer()
    #tokens = [ps.stem(token) for token in tokens if not token in stop_words]
    # Join the tokens back into a string
    clean_text = ' '.join(tokens)
    return clean_text
```

```python
inspections['Comments_Tokens'] = inspections['Comments_Tokens'].apply(lambda x: clean_text(x))
```

```python
# convert label to a binary numerical variable
inspections['Results_Flag'] = inspections.Results.map({'Fail':0, 'Pass':1})
```

```python
inspections.head(3)
```

```python
# define X and y
X = inspections.Comments_Tokens
y = inspections.Results_Flag

print(f"X Shape: {X.shape}")
print(f"y Shape: {y.shape}")
```

```python
# split X and y into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
print(f"Training records, X_train: {X_train.shape} y_train: {y_train.shape}")
print(f"Testing records, X_test: {X_test.shape} y_test: {y_test.shape}")
```

## Vectorizing our dataset

```python
# instantiate the vectorizer
vect = CountVectorizer(ngram_range= (1, 1))
```

```python
# learn training data vocabulary, then use it to create a document-term matrix
vect.fit(X_train)
X_train_dtm = vect.transform(X_train)
```

```python
# combine fit and transform into a single step
X_train_dtm = vect.fit_transform(X_train)
```

```python
# examine the document-term matrix
X_train_dtm
```

```python
# transform testing data (using fitted vocabulary) into a document-term matrix
X_test_dtm = vect.transform(X_test)
X_test_dtm
```

## Building and evaluating models

### Naive Bayes Model

```python
# instantiate a Multinomial Naive Bayes model
nb = MultinomialNB()
```

```python
# train and time the model using X_train_dtm
%time nb.fit(X_train_dtm, y_train)
```

```python
# make class predictions for X_test_dtm
y_pred_class = nb.predict(X_test_dtm)
```

```python
# calculate accuracy of class predictions
print(f"Test Accuracy: {metrics.accuracy_score(y_test, y_pred_class) * 100:.1f}%")
```

```python
# calculate precision and recall
print(classification_report(y_test, y_pred_class))
```

```python
# calculate the confusion matrix
print(metrics.confusion_matrix(y_test, y_pred_class))
```

### Logistic Regression Model

```python
# instantiate a logistic regression model
logreg = LogisticRegression(max_iter=10000)
```

```python
# train the model using X_train_dtm
%time logreg.fit(X_train_dtm, y_train)
```

```python
# make class predictions for X_test_dtm
y_pred_class = logreg.predict(X_test_dtm)
```

```python
# calculate accuracy of class predictions
print(f"Test Accuracy: {metrics.accuracy_score(y_test, y_pred_class) * 100:.1f}%")
```

```python
# calculate precision and recall
print(classification_report(y_test, y_pred_class))
```

```python
# calculate the confusion matrix
print(metrics.confusion_matrix(y_test, y_pred_class))
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV
import joblib

# Set up the pipeline
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),
    ('clf', MultinomialNB())
])

# Set up the hyperparameter grid
params = {
    'tfidf__ngram_range': [(1, 1), (1, 2), (1, 3)],
    'tfidf__max_features': [1000, 5000, None],
    'clf__alpha': [0.1, 1, 10]
}

# Set up the grid search
grid_search = GridSearchCV(pipeline, params, cv=5)

# Fit the grid search on the training data
grid_search.fit(X_train, y_train)

# Print the best parameters
print('Best parameters:', grid_search.best_params_)

# Save the model to a file
joblib.dump(grid_search.best_params_, 'nb_tfidf_model.pkl')

# Load the model from the file
loaded_model = joblib.load('nb_tfidf_model.pkl')

# Use the loaded model to make predictions
y_pred = loaded_model.predict(X_test)


# Use the best estimator to make predictions on the test data
#y_pred = grid_search.best_estimator_.predict(X_test)

# Evaluate the performance of the model on the test data
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy on test data:', accuracy)
```
